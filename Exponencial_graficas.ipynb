```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=10
lam=2
```

```python
def expo(lam=2, n=10):
  L=-(1/lam)
  lista=[]

  for i in range (n):
    u=rnd.random()
    x=L*np.log(u)
    lista.append(x)
  return lista
```

```python
expo()
```

```python
G=expo(0.5,100)
print(G)
```

```python
media=np.mean(G)
print(media)
```

```python

mediana = np.median(G)
print(mediana)
```

```python
var=np.var(G)
print(var)
```

```python
plt.hist(G)
```

```python
plt.plot(G)
```

```python
def cdf(G):
   n=len(G)
   x=np.sort(G)
   y=np.arange(0,n)/n
   return x,y
```

```python
x,y=cdf(G)
```

```python
plt.step(x,y)
plt.axvline(media, color='red', linestyle='-')
plt.axvline(mediana, color='green', linestyle='-')
```
