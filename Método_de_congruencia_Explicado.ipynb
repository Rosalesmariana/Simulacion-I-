```python
from tabulate import tabulate

x_0=17; #Elegimos la semilla inicial
a=1; #Las constantes a, c y m
c=121;
m=256;

# Encabezados de la tabla
encabezados= ["N","x_n","ax_n+c","x_n+1"]

# Almacenar resultados
resultados = []

iteration = 0

while True:
  b=a*x_0+c; #Llamamos "b" a nuestra variable que guarda el resultado de a x_n + c
  x_n=b%m; # La variable "k" llama al resultado anterior y realiza el resto de la operacion a x_n + c (mod m)

  resultados.append([iteration,x_0,b,x_n]) #Resultados que se mostraran en la tabla

# Actualizar x_0, total de iteraciones
  x_0=x_n #Ahora toma como nueva semilla al resultado anterior (x_n)
  iteration += 1

  if iteration > 260: #Por ejecuciones anteriores se sabe que el ciclo de vida es de 256, para ilistrar esto se dejan 4 ciclos mas
    break

# Tabla de resultados
print(tabulate(resultados, headers=encabezados, tablefmt="grid"))
```
