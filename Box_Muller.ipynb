```python
import numpy as np
import matplotlib.pyplot as plt

def box_muller():
    # Generar dos números aleatorios uniformemente distribuidos en el intervalo (0, 1)
    u1 = np.random.rand()
    u2 = np.random.rand()

    # Aplicar la transformación de Box-Muller
    x = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
    y = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)

    return x, y

# Generar un conjunto de números aleatorios normales
n = 10000  # Número de muestras
muestra = []

for _ in range(n // 2):
    x, y = box_muller()
    muestra.append(x)
    muestra.append(y)


muestra = np.array(muestra)

# Histograma de los números generados
plt.hist(muestra, bins=50, density=True, alpha=0.6, color='g')

# Gráfica de densidad
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = 1/np.sqrt(2 * np.pi) * np.exp(-0.5 * x**2)
plt.plot(x, p, 'k', linewidth=2)
title = "Box-Muller "
plt.title(title)
plt.show()
```
