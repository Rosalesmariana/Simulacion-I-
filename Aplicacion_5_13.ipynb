**EJEMPLO 5.13**

Un banco emplea 3 cajeros para servir a sus clientes. Los clientes arriban de acuerdo a un proceso poisson a una razón media de 40 por hora. Si un cliente encuentra todos los cajeros ocupados, entonces se incorpora a la cola que alimenta a todos los cajeros. El tiempo que dura la transacción entre un cajero y un cliente sigue una distribución uniforme entre 0 y 1 minuto. Para esta información,¿cuál es el tiempo promedio en el sistema?, ¿cuál es la cantidad promedio de clientes en el sistema?


Para abordar el problema, como primer paso identificamos que se trata sobre la *Teoria de colas*

**TEORIA DE COLAS**

Las Colas son un desequilibrio temporal cuando la demanda de un servicio es superiora la capacidad del sistema.  En la formación de colas se habla de clientes. Esta teoría estudia factores como el tiempo de espera medio en las colas o la capacidad de trabajo en el sistema sin que se llegue a colpasar.

Una Cola se presenta con frecuencia cuando se solicita un servicio por parte de una serie de clientes y tanto el servicio como los clientes son de tipo probabilístico.

DISTRIBUCIONES DE PROBABILIDAD DEL MECANISMO DE SERVICIO

Las fuentes de variación en los problemas de colas o filas de espera provienen  del carácter aleatorio de la llegada de clientes y de las variaciones que se registran en los distintos tiempos de servicio. Generalmente cada una de esas fuentes suele describirse mediante una distribución de probabilidad.

La mayor parte de los modelos de colas estocásticas asumen que el tiempo entre
diferentes llegadas de clientes siguen una distribución exponencial Exp (λ)  , o lo que es equivalente, que el ritmo de llegadas sigue una distribución de Poisson

NOTACIÓN DEL MECANISMO DE SERVICIO

$λ_n$ : Tasa media de llegadas cuando hay   clientes en el sistema, también número esperado de llegadas de clientes por unidad de tiempo cuando se encuentran clientes en el sistema


$\mu_n$ : Tasa media de servicio en todo el sistema, esto es, número esperado de clientes que son despachados por unidad de tiempo por todos los servidores en su conjunto

$c$ : Número de servidores en el sistema de colas

$U_s$ : Utilizacion promedio del sistema

$ρ$ : (factor de utilización) Congestión de un sistema

$L_s$: Número esperado de clientes en el sistema, es decir, el sumatorio de las
      probabilidades de cada estado por el número de clientes en su correspondiente
      estado

$L_q$: Longitud esperada de la cola, se trata de una variable que es medida de los
        clientes esperando en cola excluidos aquellos que están recibiendo servicio

$W_s$: Tiempo de espera en el sistema incluyendo el tiempo de servicio  para
         cada cliente.

$W_q$: Tiempo de espera en la cola excluido el tiempo de servicio  para cada
           cliente \\


David Kendall introdujo en 1953 una notación que permite describir las colas y mostrar
sus características pudiendo clasificar los diferentes tipos de colas por medio de
iniciales.

Para fines de este problema, nos encontramos frente a una cola del tipo M/G/3

$M  $ :Tiempos entre llegadas distribuidos de forma exponencial (Proceso de Poisson)


$G$ : Distribución probabilística
      de los tiempos de servicio, en este caso, "Generales", ya que, puede seguir cualquier distribución

$c=3$: Tenemos 3 servidores (cajeros)

MEDIDAS DE RENDIMIENTO:

Tiempo en el servicio:   $\frac{1}{μ}$

Utilización promedio del sistema:  $u_s= \frac{λ}{μ}$

Factor de utilización (factor de carga/ intensidad tráfico): $ρ = \frac{λ}{c*μ}$

Probabilidad de que ningún cliente
se encuentre en el sistema de colas: $P_0=\frac{1}{∑^{c-1}_{n=0}\frac{1}{n!}\frac{λ}{μ}^n + \frac{1}{s!}\frac{λ}{μ}^c \frac{1}{1-ρ}} $


Número promedio de clientes en cola: $L_q=\frac{\frac{λ}{μ}^c  λ μ}{(c-1)! (c\mu - λ)^2} * P_0$


Tiempo promedio de espera en cola: $W_q = \frac{L_q}{λ}$

Número promedio de clientes en sistema: $L_s= L_q + \frac{λ}{\mu}$

Tiempo promedio de estancia en el sistema: $W_s = W_q + \frac{1}{\mu}$


*SOLUCIÓN ANALITICA AL EJERCICIO:*

Es un modelo de cola  M/G/3

El problema nos proporciona los siguientes datos:

$λ= 40 clientes/hora $ que representa el numero de llegadas de clientes por unidad de tiempo

Para $μ$ nos indica que el tiempo que dura la transacción entre un cajero y un cliente sigue una distribución uniforme entre 0 y 1 minuto

Con esto, podemos tomar que $\mu=0.5$

Y asi $\mu=120 clientes/hora$ representa el número de clientes que son despachados por unidad de tiempo por todos los servidores

Una vez establecido esto podemos determinar el número promedio de clientes en el sistema de la siguiente manera

Consideramos

$ρ= \frac{λ}{c μ} = \frac{40}{3 x 120} =\frac{1}{9}$

$U_s=\frac{\lambda}{\mu}= \frac{40}{120} =\frac{1}{3}$


$P_0=\frac{1}{∑^{c-1}_{n=0}\frac{1}{n!}\frac{λ}{μ}^n + \frac{1}{s!}\frac{λ}{μ}^c \frac{1}{1-ρ}} $

$P_0=\frac{1}{∑^{2}_{n=0}\frac{1}{n!}\frac{40}{120}^n + \frac{1}{3!}\frac{40}{120}^3 \frac{1}{1-\frac{1}{9}}} $

$P0=0.7438$

Asi, el el número promedio de clientes es

$L_s = L_q + (\frac{λ}{\mu})$

donde

$L_q = \frac{u_s^2 \lambda \mu }{2! * (3*\mu - \lambda)^2}*P_0$

$L_q= 0.00193$

$⇒ L_s= 0.00193 + \frac{40}{120} =  0.33 $ clientes

Luego, para calcular el tiempo promedio en el sistema
$W_s= W_q + \frac{1}{\mu}*60$

Donde,

$W_q=\frac{L_q}{λ}$

$W_q=\frac{0.00193}{40}=4.84295e-05$

Asi

$W_s= 4.84295e-05 + \frac{1}{120}*60 = 0.50 minutos$

Con esto, hemos llegado al final de la solucion analitica

*SOLUCION POR SIMULACION*

Para hallar la solucion de esta problema sobre teoria de colas, se implemento el siguiente código


```python
import math
import random as rnd

# Parámetros del sistema
LAMBDA = 40  # tasa de llegada (clientes por minuto)
MU=120   # tasa de servicio
c=3 # numero de cajeros
n=0

# Calcular rho
rho = LAMBDA / (c*MU)
# Calcular la utilización  promedio del sistema
u_s=LAMBDA/MU


# Calcular P0 usando la aproximación para M/G/3
def calcular_P0(u_s, c, rho,n):
    sumatoria = sum([((u_s ** n) / math.factorial(n)) + ((u_s**c) * (1/math.factorial(3))* (1/1-rho)) for n in range(c-1)] )
    denominador = sumatoria
    return 1 / denominador

P0 = calcular_P0(u_s, c, rho,n) # Probabilidad de que ningún cliente se encuentre en el sistema de colas

# Calcular número promedio de clientes en el sistema
L_q = ((u_s**2 *LAMBDA*MU)/ ((math.factorial(c-1))*(c*MU-LAMBDA)**2)) * (P0)
L_s = L_q + (LAMBDA/MU)

# Calcular el tiempo promedio en el sistema
W_q= L_q/LAMBDA
W_s= (W_q + (1/MU))*60


print(f'Número promedio de clientes en el sistema : {L_s:.2f} clientes')
print(f'Tiempo promedio en el sistema: {W_s:.2f} minutos')
```

La implementacion del codigo siguiente, se debe a el cambio que puede tener $\mu$ ya que en el problema indica que, el tiempo que dura la transacción entre un cajero y un cliente sigue una distribución uniforme entre 0 y 1 minuto.

Para ello el valor de $\mu$ para este codigo toma un valor aletorio que sigue una distribucion unifirme entre $0$ y $1$

```python
import math
import random as rnd

# Parámetros del sistema
LAMBDA = 40  # tasa de llegada (clientes por minuto)
MU = 60/(rnd.uniform(0, 1)) # tasa de servicio promedio (clientes por minuto por cajero)
c=3 #numero de cajeros


# Calcular rho
rho = LAMBDA / (c*MU)
# Calcular la utilización  promedio del sistema
u_s=LAMBDA/MU


# Calcular P0 usando la aproximación para M/M/3
def calcular_P0(u_s, c, rho,n):
    sumatoria = sum([((u_s ** n) / math.factorial(n)) + ((u_s**c) * (1/math.factorial(3))* (1/1-rho)) for n in range(c-1)] )
    denominador = sumatoria
    return 1 / denominador

P0 = calcular_P0(u_s, c, rho,n) # Probabilidad de que ningún cliente se encuentre en el sistema de colas

# Calcular número promedio de clientes en el sistema
L_q = ((u_s**2 *LAMBDA*MU)/ ((math.factorial(c-1))*(c*MU-LAMBDA)**2)) * (P0)
L_s = L_q + (LAMBDA/MU)

# Calcular el tiempo promedio en el sistema
W_q= L_q/LAMBDA
W_s= (W_q + (1/MU))*60


print(f'Número promedio de clientes en el sistema : {L_s:.2f} clientes')
print(f'Tiempo promedio en el sistema: {W_s:.2f} minutos')
```
