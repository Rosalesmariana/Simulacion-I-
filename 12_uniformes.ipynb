```python
import random
import matplotlib.pyplot as plt

def U():
    U=[]
    for i in range(12):
        u=random.uniform(0,1)
        U.append(u)
    s=sum(U)
    return s

X=[]
for i in range(10000):
    x=U()-6
    X.append(x)

plt.hist(X,bins=500)
plt.show()
```
